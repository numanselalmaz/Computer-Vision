```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

def imshow(title = "Image", image = None, size = 10):
    w, h = image.shape[0], image.shape[1]
    aspect_ratio = w/h
    plt.figure(figsize=(size * aspect_ratio,size))
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.show()
```

cv2.findContours işlemi, OpenCV (Open Source Computer Vision Library) kütüphanesinde bulunan ve görüntü işleme uygulamalarında nesnelerin veya yapıların konturlarını (dış çizgilerini) tespit etmek için kullanılan bir fonksiyondur. Bu işlem, görüntü işleme projelerinde nesne algılama, nesne sayımı, nesne izleme, görüntü segmentasyonu ve daha birçok uygulama için önemlidir.

cv2.findContours işlemi şu şekilde kullanılır:

python
Copy code
contours, hierarchy = cv2.findContours(image, mode, method)
Bu işlemin parametreleri aşağıdaki gibidir:

image: Konturların tespit edileceği görüntü. Genellikle siyah-beyaz (binary) bir görüntüdür, yani nesneler siyah, arka plan beyaz olarak temsil edilir.

mode: Kontur algılama modunu belirler. Bu parametre, konturların nasıl algılanacağını kontrol eder. Örnek modlar şunlar olabilir:

cv2.RETR_EXTERNAL: Yalnızca dış konturları bulur.
cv2.RETR_LIST: Tüm konturları bulur ve hiyerarşi kullanılmaz.
cv2.RETR_CCOMP: Tüm konturları iki seviyeli bir hiyerarşi olarak bulur.
cv2.RETR_TREE: Tüm konturları tam bir hiyerarşi olarak bulur.
method: Kontur yaklaşım yöntemini belirler. Bu parametre, konturun dış çizgisi veya iç çizgisi gibi nasıl temsil edileceğini kontrol eder. Örnek yöntemler şunlar olabilir:

cv2.CHAIN_APPROX_NONE: Tüm kontur noktalarını korur.
cv2.CHAIN_APPROX_SIMPLE: Yalnızca konturun başlangıç ve bitiş noktalarını saklar.
cv2.CHAIN_APPROX_TC89_L1 ve cv2.CHAIN_APPROX_TC89_KCOS: Tehlike tespiti işlemleri için özel yaklaşımlar.


```python
image = cv2.imread('images/LP.jpg')

gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

_, th2 = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
imshow('After thresholding', th2)

contours, hierarchy = cv2.findContours(th2, cv2.RETR_LIST, cv2.CHAIN_APPROX_NONE)

cv2.drawContours(image, contours, -1, (0, 255,0), thickness = 2)
imshow('Contours overlaid on original image', image)

print('Number of Contours found = ' + str(len(contours)))

print(hierarchy)
```

### ** RETR_LIST ** 
Tüm konturları alır ancak herhangi bir üst-alt öğe ilişkisi oluşturmaz. Bu kurala göre ebeveynler ve çocuklar eşittir ve onlar sadece birer sınırdır. yani hepsi aynı hiyerarşi düzeyine aittir.

```python
image = cv2.imread('images/LP.jpg')

gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
_, th2 = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
imshow('After thresholding', th2)

contours, hierarchy = cv2.findContours(th2, cv2.RETR_LIST, cv2.CHAIN_APPROX_NONE)

cv2.drawContours(image, contours, -1, (0,255,0), thickness = 2)
imshow('Contours overlaid on original image', image)

print("Number of Contours found = " + str(len(contours)))
print(hierarchy)
```

### **RETR_EXTERNAL**

Yalnızca aşırı dış bayrakları döndürür. Tüm alt konturlar geride bırakılır.

```python
image = cv2.imread('images/LP.jpg')

gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

_, th2 = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
imshow('After thresholding', th2)

contours, hierarchy = cv2.findContours(th2, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)

cv2.drawContours(image, contours, -1, (0,255,0), thickness = 2)
imshow('Contours overlaid on original image', image, size = 10)

print("Number of Contours found = " + str(len(contours)))
print(hierarchy)
```

### **RETR_CCOMP**

Tüm konturları alır ve bunları 2 seviyeli bir hiyerarşiye göre düzenler. yani nesnenin dış hatları (yani sınırları) hiyerarşi-1'e yerleştirilir. Ve nesnenin içindeki deliklerin dış hatları (varsa) hiyerarşi-2'ye yerleştirilir. İçinde herhangi bir nesne varsa konturu yine yalnızca hiyerarşi-1'e yerleştirilir. Ve hiyerarşi-2'deki delik vb.

```python
image = cv2.imread('images/LP.jpg')

gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

_, th2 = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
imshow('After thresholding', th2)

contours, hierarchy = cv2.findContours(th2, cv2.RETR_CCOMP, cv2.CHAIN_APPROX_NONE)


cv2.drawContours(image, contours, -1, (0,255,0), thickness = 2)
imshow('Contours overlaid on original image', image)

print("Number of Contours found = " + str(len(contours)))
print(hierarchy)
```
