```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

def imshow(title = "Image", image = None, size = 10):
    w, h = image.shape[0], image.shape[1]
    aspect_ratio = w/h
    plt.figure(figsize=(size * aspect_ratio,size))
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.show()
```

```python
image = cv2.imread('images/liberty.jpeg', 0) #original not equal 0, equal 1 
imshow("Grayscaled Image",image)
print(image)
image1 = cv2.imread('images/liberty.jpeg', 1) #original not equal 0, equal 1
imshow('Original', image1)
print(image1)
# check print(image) and print(image1) outputs. different for arithmetic.
```

```python
# Create a matrix of ones, then multiply it by a scaler of 100 
# This gives a matrix with same dimesions of our image with all values being 100
M = np.ones(image.shape, dtype = "uint8") * 100 
M
```

# Increasing Brightness

```python
added = cv2.add(image, M)
imshow("Increasing Brightness", added)

added2 = image + M 
imshow("Simple Numpy Adding Results in Clipping", added2)

print("added:",added)
print("added2:",added2)
"""
why i am setting brightness in classic with add function my add operation produces different result ?
- 
OpenCV's add()function compares the differences between the pixels of the images and uses a threshold value to fix the value of the pixels 255to or . 0This is a very useful method for adjusting the brightness of the image.

Classic Python addition performs a normal addition between two numbers. This is often not enough to adjust the brightness of the image, because the values ​​of the pixels may exceed 255 or be less than 0.
"""
```

```python
subtracted = cv2.subtract(image, M)
imshow("Subtracted", subtracted)

subtracted = image - M 
imshow("Subtracted 2", subtracted)
```

# Bitwise Opertaions and Masking

```python
square = np.zeros((300, 300), np.uint8)
cv2.rectangle(square, (50, 50), (250, 250), 255, -2)
imshow("square", square)

ellipse = np.zeros((300, 300), np.uint8)
cv2.ellipse(ellipse, (150, 150), (150, 150), 30, 0, 180, 255, -1)
imshow("ellipse", ellipse)
```

bitwise_not : The function can be used to invert the image, thresholding or masking.

bitwise_xor : function in OpenCV performs a bitwise XOR operation on two images or arrays. The bitwise XOR operation is a logical operation that returns a 1 if and only if the corresponding bits of the two inputs are different.

bitwise_or : function in OpenCV performs a bitwise OR operation on two images or arrays. The bitwise OR operation is a logical operation that returns a 1 if either of the corresponding bits of the two inputs is 1.

bitwise_and : function in Python performs a bitwise AND operation on two arrays. The bitwise AND operation is a logical operation that returns a 1 if and only if the corresponding bits of the two inputs are both 1.

```python
And = cv2.bitwise_and(square, ellipse)
imshow("AND", And)

bitwiseOr = cv2.bitwise_or(square, ellipse)
imshow("bitwiseOr", bitwiseOr)

bitwiseXor = cv2.bitwise_xor(square, ellipse)
imshow("bitwiseXor", bitwiseXor)

bitwiseNot_sq = cv2.bitwise_not(ellipse)
imshow("bitwiseNot_sq", bitwiseNot_sq)

```
